# Apply the t-test to real data

Your assignment is to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

Your goals:

1. Load and clean the data (or determine the best method to drop observations when running tests)
2. Using hypothesis testing, find an issue that democrats support more than republicans with p < 0.01
3. Using hypothesis testing, find an issue that republicans support more than democrats with p < 0.01
4. Using hypothesis testing, find an issue where the difference between republicans and democrats has p > 0.1 (i.e. there may not be much of a difference)

Note that this data will involve *2 sample* t-tests, because you're comparing averages across two groups (republicans and democrats) rather than a single group against a null hypothesis.

```python
### YOUR CODE STARTS HERE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import *

!wget https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data --no-check-certificate
```

```python
#setting the column headers to a list of values for the issues they voted
column_headers = ['party', 'handicapped-infants', 'water-project', 'budget',
             'physician-fee-freeze', 'el-salvador-aid', 'religious-groups',
             'anti-satallite-ban', 'aid-to-contras', 'mx-missile', 
             'immigration', 'synfuels', 'education', 'right-to-sue',
             'crime', 'duty-free', 'south-africa']
#reading in the data setting the header to none, and specifying the names of the column headers, as well as specifying na_values
houseDf = pd.read_csv("/content/house-votes-84.data",
                      header=None,
                      names=column_headers,
                      na_values='?')

#inspecting the head, and tail of the dataframe
houseDf
```

```python
#replaying the characters in dataframe with numeric values to get vote counts
houseDf = houseDf.replace({'y':1, 'n':0})
houseDf
```

```python
#setting a df to hold the republican vote count on issues, and checking the shape to make sure we dont have extra columns
gop = houseDf[houseDf['party']=='republican']
gop.shape
```

```python
#setting a df to hold the democratic vote count on issues, and checking the shape to make sure we dont have extra columns
dnc = houseDf[houseDf['party']=='democrat']
dnc.shape
```

```python
#itterating through the columns in the dataframe, and using tuple unpacking to help find issues where the democtratic vote was greater than the republican vote within a certain pvalue
for vote in dnc.columns[1:]:
  statistic,pvalue = ttest_ind(dnc[vote], gop[vote], nan_policy='omit')
  if pvalue <.01:
    if statistic > 0:
      print("democrats supported " f'vote:{vote}' " more")
      print(f'vote:{vote} statistic:{statistic:.5f} pvalue:{pvalue:.5e}')
```

```python
#itterating through the columns in the dataframe, and using tuple unpacking to help find issues where the republican vote was greater than the democtratic vote within a certain pvalue
for col in gop.columns[1:]:
  statistic,pvalue = ttest_ind(gop[col], dnc[col], nan_policy='omit')
  #print(col, statistic,pvalue)
  if pvalue < .01:
    if statistic > 0:
      print("Republicans supported " f'col:{col}' " more")
      print(f'col:{col} statistic:{statistic:.5f} pvalue:{pvalue:.5e}')
```

```python
#itterating through the issues, looking for an issue where the democratic, and rupublican support with a pvalue >.1
for col in gop.columns[1:]:
  statistic,pvalue = ttest_ind(gop[col], dnc[col], nan_policy='omit')
  #print(col, statistic,pvalue)
  if pvalue > .1:
    print("the support for " f'col:{col}' " is very close in value for both groups with a " f'pvalue:{pvalue}')
    
```
